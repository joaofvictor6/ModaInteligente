#  Bem-vindo √† An√°lise Explorat√≥ria dos Produtos de Vestu√°rio!

O objetivo aqui √© explorar o *Fashion Clothing Products Dataset* para entender melhor as caracter√≠sticas dos produtos e como esses dados podem ser usados para criar um sistema de recomenda√ß√£o eficiente.


```python
# Instalar a biblioteca kagglehub, caso n√£o esteja instalada
!pip install kagglehub

# Importando as bibliotecas necess√°rias
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo do seaborn
sns.set(style="whitegrid")

# Baixar o dataset diretamente do Kaggle
path = kagglehub.dataset_download("shivamb/fashion-clothing-products-catalog")

# Carregar o dataset baixado
dataset_path = f'{path}/myntra_products_catalog.csv'
df = pd.read_csv(dataset_path)
```

## üîç Informa√ß√µes B√°sicas sobre o Dataset

Antes de come√ßarmos, vamos dar uma olhada nas informa√ß√µes b√°sicas do dataset, como a quantidade de registros, colunas e tipos de dados.


```python
# Exibir informa√ß√µes gerais sobre o dataset
df.info()

# Exibir as primeiras linhas do dataset
df.head()
```

## üí∏ An√°lise 1: Distribui√ß√£o dos Pre√ßos dos Produtos

A primeira an√°lise que faremos √© sobre a distribui√ß√£o dos pre√ßos dos produtos. Isso nos ajudar√° a entender se h√° uma grande varia√ß√£o nos pre√ßos e se o dataset cont√©m mais produtos de alto ou baixo valor.


```python
# Distribui√ß√£o dos Pre√ßos
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (INR)'], bins=30, kde=True)
plt.title('Distribui√ß√£o dos Pre√ßos dos Produtos (INR)', fontsize=14)
plt.xlabel('Pre√ßo (INR)', fontsize=12)
plt.ylabel('Frequ√™ncia', fontsize=12)
plt.show()
```

**üí° Insights**: A distribui√ß√£o de pre√ßos mostra que a maioria dos produtos tem pre√ßos mais acess√≠veis, mas tamb√©m existem alguns produtos de alto valor, sugerindo uma oferta diversa para diferentes perfis de consumidores.


## üëóüëî An√°lise 2: Distribui√ß√£o dos Produtos por G√™nero

Agora, vamos explorar a distribui√ß√£o dos produtos por g√™nero. Com isso, podemos ver se a cole√ß√£o √© predominantemente voltada para algum p√∫blico em espec√≠fico ou se h√° um equil√≠brio entre os g√™neros.


```python
# Distribui√ß√£o por G√™nero
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title('Distribui√ß√£o dos Produtos por G√™nero', fontsize=14)
plt.xlabel('G√™nero', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.show()
```

**üí° Insights**: Observamos que a maioria dos produtos √© voltada para o p√∫blico feminino, o que pode indicar um foco maior nesse segmento. No entanto, tamb√©m h√° uma boa variedade de produtos masculinos e unissex.


## üè∑Ô∏è An√°lise 3: As 10 Marcas Mais Frequentes

Al√©m de entender o p√∫blico-alvo, √© importante conhecer as marcas mais representadas na nossa base de dados. Vamos agora verificar quais s√£o as 10 marcas mais frequentes e com mais produtos dispon√≠veis.


```python
# Marcas Mais Frequentes
plt.figure(figsize=(12, 6))
top_brands = df['ProductBrand'].value_counts().nlargest(10)
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 Marcas Mais Frequentes', fontsize=14)
plt.xlabel('Marca do Produto', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

**üí° Insights**: As marcas mais frequentes incluem nomes populares como 'Roadster', 'HRX' e 'DressBerry'. Esse dado √© √∫til para entender quais marcas est√£o mais presentes no cat√°logo e podem ser mais relevantes para o sistema de recomenda√ß√£o.
