```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

sns.set(style="whitegrid")
```

```python
# Carregar o dataset
df = pd.read_csv("../dados/myntra_products_catalog.csv")

# Informações gerais do dataset
df.info()

# Visualizar as primeiras linhas
df.head()
```

```python
# Visualização de Valores Ausentes com Melhorias
import missingno as msno
import matplotlib.pyplot as plt

# Configuração do estilo do Matplotlib
plt.style.use("ggplot")

# Criando a figura
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Gráfico de matriz de valores ausentes
msno.matrix(df, ax=axes[0], fontsize=12, sparkline=False)
axes[0].set_title("Visualização dos Valores Ausentes por Linha e Coluna", fontsize=16)
axes[0].set_xlabel("Colunas", fontsize=12)
axes[0].set_ylabel("Linhas (Registros)", fontsize=12)

# Gráfico de barras de valores ausentes
msno.bar(df, ax=axes[1], fontsize=12, color="steelblue")
axes[1].set_title("Proporção de Valores Ausentes por Coluna", fontsize=16)
axes[1].set_xlabel("Colunas", fontsize=12)
axes[1].set_ylabel("Proporção de Valores", fontsize=12)

# Rotação dos rótulos do eixo X
for ax in axes:
    ax.tick_params(axis='x', labelrotation=45)

# Ajustando o layout
plt.tight_layout()
plt.show()
```

```python
# Distribuição dos preços
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (INR)'], bins=30, kde=True)
plt.title("Distribuição de Preços")
plt.xlabel("Preço (INR)")
plt.ylabel("Frequência")
plt.show()

# Produtos por gênero
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title("Distribuição de Produtos por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Contagem")
plt.show()

# Marcas mais frequentes
plt.figure(figsize=(12, 6))
top_brands = df['ProductBrand'].value_counts().nlargest(10)
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title("Top 10 Marcas Mais Frequentes")
plt.xlabel("Marca")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```
